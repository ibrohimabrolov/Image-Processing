```python
# ==============================
# LAB 5 - TASK 1
# Histogram Equalization (Grayscale)
# ==============================

# 1️⃣ Install required libraries (Colab uchun)
!pip install opencv-python matplotlib numpy scikit-image

# 2️⃣ Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# 3️⃣ Upload image
uploaded = files.upload()

# 4️⃣ Read uploaded image as grayscale
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if image loaded correctly
if image is None:
    print("Image not loaded! Please upload again.")
else:
    print("Image successfully loaded!")

# 5️⃣ Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)

# 6️⃣ Display Original & Equalized images with histograms
plt.figure(figsize=(12,8))

# ---- Original Image ----
plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# ---- Original Histogram ----
plt.subplot(2,2,2)
plt.hist(image.ravel(), bins=256, range=[0,256])
plt.title("Histogram (Original)")

# ---- Equalized Image ----
plt.subplot(2,2,3)
plt.imshow(equalized_image, cmap='gray')
plt.title("Equalized Image")
plt.axis("off")

# ---- Equalized Histogram ----
plt.subplot(2,2,4)
plt.hist(equalized_image.ravel(), bins=256, range=[0,256])
plt.title("Histogram (Equalized)")

plt.tight_layout()
plt.show()
```

```python
# ==============================
# LAB 5 - TASK 2
# Histogram Equalization on Color Image
# ==============================

# 1️⃣ Import libraries (agar oldin import qilmagan bo‘lsang)
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# 2️⃣ Upload color image
uploaded = files.upload()

# 3️⃣ Read image (COLOR mode)
image_path = list(uploaded.keys())[0]
color_image = cv2.imread(image_path)

# Check image
if color_image is None:
    print("Image not loaded!")
else:
    print("Color image loaded successfully!")

# 4️⃣ Split into B, G, R channels
b, g, r = cv2.split(color_image)

# 5️⃣ Apply histogram equalization to each channel separately
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)

# 6️⃣ Merge equalized channels back
equalized_color_image = cv2.merge([
    equalized_b,
    equalized_g,
    equalized_r
])

# 7️⃣ Display results
plt.figure(figsize=(12,6))

# Original
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title("Original Color Image")
plt.axis("off")

# Equalized
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(equalized_color_image, cv2.COLOR_BGR2RGB))
plt.title("Equalized Color Image (Channels Separately)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# ==============================
# LAB 5 - TASK 3
# Histogram Matching
# ==============================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
from google.colab import files

# 🔹 Upload TWO images AT ONCE
print("Upload SOURCE and TARGET images together")
uploaded = files.upload()

# Get file names
file_names = list(uploaded.keys())

if len(file_names) < 2:
    print("Please upload TWO images!")
else:
    source_path = file_names[0]
    target_path = file_names[1]

    # Read images
    image_source = cv2.imread(source_path, cv2.IMREAD_GRAYSCALE)
    image_target = cv2.imread(target_path, cv2.IMREAD_GRAYSCALE)

    # Histogram Matching
    matched_image = exposure.match_histograms(image_source, image_target)

    matched_image = np.clip(matched_image, 0, 255).astype(np.uint8)

    # Display
    plt.figure(figsize=(15,8))

    # Source
    plt.subplot(3,3,1)
    plt.imshow(image_source, cmap='gray')
    plt.title("Source")
    plt.axis("off")

    plt.subplot(3,3,2)
    plt.hist(image_source.ravel(), 256, [0,256])
    plt.title("Histogram (Source)")

    # Target
    plt.subplot(3,3,4)
    plt.imshow(image_target, cmap='gray')
    plt.title("Target")
    plt.axis("off")

    plt.subplot(3,3,5)
    plt.hist(image_target.ravel(), 256, [0,256])
    plt.title("Histogram (Target)")

    # Matched
    plt.subplot(3,3,7)
    plt.imshow(matched_image, cmap='gray')
    plt.title("Matched")
    plt.axis("off")

    plt.subplot(3,3,8)
    plt.hist(matched_image.ravel(), 256, [0,256])
    plt.title("Histogram (Matched)")

    plt.tight_layout()
    plt.show()
```

```python
# ==============================
# LAB 5 - TASK 4
# Equalization + Matching
# ==============================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
from google.colab import files

# 1️⃣ Upload TWO grayscale images
print("Upload SOURCE image")
uploaded1 = files.upload()
source_path = list(uploaded1.keys())[0]

print("Upload TARGET image")
uploaded2 = files.upload()
target_path = list(uploaded2.keys())[0]

# 2️⃣ Read images
source = cv2.imread(source_path, cv2.IMREAD_GRAYSCALE)
target = cv2.imread(target_path, cv2.IMREAD_GRAYSCALE)

# 3️⃣ Step 1: Equalize source
equalized_source = cv2.equalizeHist(source)

# 4️⃣ Step 2: Match equalized image to target
matched_after_eq = exposure.match_histograms(equalized_source, target)
matched_after_eq = np.clip(matched_after_eq, 0, 255).astype(np.uint8)

# 5️⃣ Display
plt.figure(figsize=(15,10))

# Original
plt.subplot(2,3,1)
plt.imshow(source, cmap='gray')
plt.title("Original Source")
plt.axis("off")

# Equalized
plt.subplot(2,3,2)
plt.imshow(equalized_source, cmap='gray')
plt.title("Equalized Source")
plt.axis("off")

# Final Matched
plt.subplot(2,3,3)
plt.imshow(matched_after_eq, cmap='gray')
plt.title("Equalized + Matched")
plt.axis("off")

# Histograms
plt.subplot(2,3,4)
plt.hist(source.ravel(), 256, [0,256])
plt.title("Histogram (Original)")

plt.subplot(2,3,5)
plt.hist(equalized_source.ravel(), 256, [0,256])
plt.title("Histogram (Equalized)")

plt.subplot(2,3,6)
plt.hist(matched_after_eq.ravel(), 256, [0,256])
plt.title("Histogram (Final)")

plt.tight_layout()
plt.show()
```
